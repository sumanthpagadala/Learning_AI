```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn import datasets

iris = datasets.load_iris()
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                  columns= iris['feature_names'] + ['target'])
df.head()
```

** Create a pairplot of the data set. Which flower species seems to be the most separable?**

```python
sns.pairplot(df, palette='Dark2')
```

```python
sns.kdeplot(df['sepal length (cm)'], cmap="plasma", shade=True, shade_lowest=False)
```

```python
sns.kdeplot(df['sepal width (cm)'], cmap="plasma", shade=True, shade_lowest=False)

sns.kdeplot(df['petal length (cm)'], cmap="plasma", shade=True, shade_lowest=False)

sns.kdeplot(df['petal width (cm)'], cmap="plasma", shade=True, shade_lowest=False)

sns.jointplot(x='sepal length (cm)',y='sepal width (cm)',data=df,kind='kde',color='blue')

sns.jointplot(x='petal length (cm)',y='petal width (cm)',data=df,kind='kde',color='blue')

sns.jointplot(x='sepal length (cm)',y='petal length (cm)',data=df,kind='kde',color='blue')

sns.jointplot(x='sepal width (cm)',y='petal width (cm)',data=df,kind='kde',color='blue')

sns.jointplot(x='sepal length (cm)',y='petal width (cm)',data=df,kind='kde',color='blue')

sns.jointplot(x='sepal width (cm)',y='petal length (cm)',data=df,kind='kde',color='blue')

sns.jointplot(x='sepal length (cm)',y='petal length (cm)',data=df,kind='kde',color='blue')

sns.jointplot(x='sepal width (cm)',y='petal width (cm)',data=df,kind='kde',color='blue')

sns.pairplot(df,hue='target',palette='Dark2')
```

```python

from sklearn.model_selection import train_test_split
X = df.drop('target',axis=1)
y = df['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
from sklearn.svm import SVC
svc_model = SVC()
svc_model.fit(X_train,y_train)

predictions = svc_model.predict(X_test)

```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,predictions))
print(classification_report(y_test,predictions))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C':[0.1,1,10,100,1000],'gamma':[1,0.1,0.001,0.0001]}
grid = GridSearchCV(SVC(),param_grid,verbose=3)
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid_predictions = grid.predict(X_test)
print(confusion_matrix(y_test,grid_predictions))
print(classification_report(y_test,grid_predictions))
```
