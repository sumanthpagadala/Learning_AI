```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#object oriented method
fig = plt.figure() #imaginary blank canvas

axes = fig.add_axes([0.1,0.1,0.8,0.8])
# left bottom, width and height

axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)

axes2.plot(y,x,'r')
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2)

plt.tight_layout() # to remove over lapping of numbers

axes[0].plot(x,y)
axes[1].plot(y,x,'b')
```
